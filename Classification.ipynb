```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cross_validation import train_test_split
```

### Step 1  数据集导入

```python
data_2 = pd.read_csv('iBeacon_RSSI_Labeled.csv')
```

```python
data_2.head(5)
```

```python
print(data_2.info())
```

### Step 2  数据预处理

###### 2.1 删除无用列,替换无效值

```python
#删除‘日期’列
del data_2['date']
```

```python
print(data_2.info())
```

```python
data_2.replace(-200,0,inplace = True)
```

```python
data_2.head(10)
```

###### 2.2 统计数据集中出现的不同 location 数目

```python
loc_count={}
for i in data_2['location'].index:
    if data_2['location'][i] in loc_count:
        loc_count[data_2['location'][i]] += 1
    else:
        loc_count[data_2['location'][i]] = 1 

print(len(loc_count))
```

```python
data_2.head(20)
```

```python
print(loc_count)
```

###### 2.3 将105个位置映射成数字，方便模型进行数据处理

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(data_2['location'])
data_2['location'] = le.transform(data_2['location'])
```

```python
data_2.head(5)
```

### Step 3 将数据集分成训练集和测试集

```python
features = data_2
```

```python
n_features = features.drop('location', axis=1)
n_samples = features['location']
```

```python
X_train, X_test,y_train,y_test = train_test_split(n_features,n_samples,test_size=0.2)
```

### Step 4  运用算法实现多分类问题，预测location的位置

### Way-1: SVM (Multi-class)

```python
from sklearn import svm
clf = svm.SVC(gamma='auto', decision_function_shape='ovo')
clf.fit(X_train, y_train) 
```

```python
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
#查看准确度Accuracy,Precission,Recall
from sklearn import metrics

accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred,average='micro')
recall = metrics.recall_score(y_test, y_pred,average='micro')
```

```python
print("SVM:the accuracy is: ",accuracy)
print("SVM:the precision is: ",precision)
print("SVM:the recall is: ",recall)
```

### Way-2: Decision Tree

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics

accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred,average='micro')
recall = metrics.recall_score(y_test, y_pred,average='micro')
```

```python
print("Decision Tree:the accuracy is: ",accuracy)
print("Decision Tree:the precision is: ",precision)
print("Decision Tree:the recall is: ",recall)
```

### Way-3: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
clf = RandomForestClassifier(n_estimators=100, max_depth=10,
                             random_state=1)
clf.fit(X_train, y_train)  
```

```python
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics

accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred,average='micro')
recall = metrics.recall_score(y_test, y_pred,average='micro')
```

```python
print("Random Tree:the accuracy is: ",accuracy)
print("Random Tree:the precision is: ",precision)
print("Random Tree:the recall is: ",recall)
```
